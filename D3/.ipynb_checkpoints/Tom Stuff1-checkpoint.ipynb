# Import & Preprocessing

```python
import pandas as pd
import numpy as np
import datetime

feedbacks = pd.read_csv(r'./feedbacks.csv')
items = pd.read_csv(r'./items.csv')
items.set_index(keys="item_hash", inplace=True)

#feedbacks preprocessing
feedbacks['date'] = feedbacks['date'].apply(lambda x: pd.to_datetime(x).year)
feedbacks = feedbacks[~feedbacks['category'].str.contains("other")]
feedbacks.sort_values(by="date", inplace=True)
```

```python
#per item hash get the category
categoryPerItem = dict.fromkeys(feedbacks["item_hash"].unique())

for item in categoryPerItem.keys():
    categoryPerItem[item] = items.loc[item].category
```

# Items per year

```python
itemsPerYear= {}
for year in range(2011, 2018):
    itemsPerYear[year] = feedbacks[feedbacks["date"].isin([year])].set_index(keys="hash_str").drop(labels=["receiver_hash", "message", "giver_hash", "order_title", "feedback_value", "norm_feedback_value", "norm_order_value"], axis=1)
```

```python
#itemsPerYear[2011]
#items["marketplace"].unique()
feedbacks["marketplace"].unique()
```

# Getting the sums

```python
for year in range(2011, 2018):
    print("year: " + str(year))
    for market in feedbacks["marketplace"].unique():
        print("  market: " + str(market))
        marketItems = itemsPerYear[year][itemsPerYear[year]["marketplace"] == market]
        total = marketItems.sum()["order_amount_usd"]
        print("  total sum: " + str(total))
        #cashOutTotal = itemsPerYear[year][(itemsPerYear[year]["category"] == "cash-out"].sum()["order_amount_usd"]
        cashOutTotal = marketItems[marketItems["category"] == "cash-out"].sum()["order_amount_usd"]
        print("  cash-out total: " + str(cashOutTotal))
        print("  ratio: " + "{:.2f}".format(cashOutTotal/total))
        print("-----------------------------------------------------------")
    print("===============================================================")
```
